**Binomial/Bernoulli process**: This is the Memory less process where each of the trials are independent of what happened in the past. The probability of success and failure remains constant throughout the process.<br>
$P(Success) = p $ <br> 
$P(failure) = (1-p)$ <br> 
Probability of getting k success out of n trials will be given by:<br>
$p^k (1 - p)^{n-k}$ <br> 
This gives one set of arrangements for Success and Failure. Using the permutaion and combination for having k-successes we come up with the binomial distribution as: <br>
\begin{equation}
f(x) = \binom{n}{k}p^k (1 - p)^{n-k}
\end{equation}

The distribution characteristics are: <br>
mean, $\mu$ = $np$ <br>
variance, $\sigma^2$ = $np(1-p)$ <br>

Let's draw a binomial distribution for varying probabilities of success in 11 trials:

```python
x = scipy.linspace(0,10,11) #Return evenly spaced numbers over a specified interval.

pmf_10 = scipy.stats.binom.pmf(x,10,0.1)
pmf_30 = scipy.stats.binom.pmf(x,10,0.3)
pmf_50 = scipy.stats.binom.pmf(x,10,0.5)
pmf_70 = scipy.stats.binom.pmf(x,10,0.7)
pmf_90 = scipy.stats.binom.pmf(x,10,0.9)
pmf_100 = scipy.stats.binom.pmf(x,10,1)

plt.figure(figsize=(12, 8))

plt.subplot(231)
plt.scatter(x,pmf_10)
plt.xlim(0,10)
plt.ylim(0,0.5)
plt.stem(x,pmf_10, 'b')
plt.title("p = 0.1")

plt.subplot(232)
plt.scatter(x,pmf_30)
plt.xlim(0,10)
plt.ylim(0,0.5)
plt.stem(x,pmf_30, 'b')
plt.title("p = 0.3")

plt.subplot(233)
plt.scatter(x,pmf_50)
plt.xlim(0,10)
plt.ylim(0,0.5)
plt.stem(x,pmf_50, 'b')
plt.title("p = 0.5")

plt.subplot(234)
plt.scatter(x,pmf_70)
plt.xlim(0,10)
plt.ylim(0,0.5)
plt.stem(x,pmf_70, 'b')
plt.title("p = 0.7")

plt.subplot(235)
plt.scatter(x,pmf_90)
plt.xlim(0,10)
plt.ylim(0,0.5)
plt.stem(x,pmf_90, 'b')
plt.title("p = 0.9")

plt.subplot(236)
plt.scatter(x,pmf_100)
plt.xlim(0,100)
plt.ylim(0,0.5)
plt.stem(x,pmf_100, 'b')
plt.title("p = 1")

plt.show()
```

Look at how beautifully the graph moves from being skewed to a more symmetric as the probability of success and failure comes closer..
